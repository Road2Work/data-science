```python
def get_static_questions(target_role, role_family):
    """
    Menghasilkan pertanyaan statis (HR, Behavioral, Closing) untuk posisi target_role dalam role_family tertentu. 
    Pertanyaan ini dirancang untuk mengevaluasi kompetensi seperti motivasi karir, teamwork, adaptabilitas, dan kesadaran diri. 
    Setiap pertanyaan dilengkapi dengan guardrail signals yang membantu menilai kualitas jawaban kandidat berdasarkan aspek-aspek yang relevan dengan posisi yang dilamar.
    """
    return [
        # ==========================================
        # KATEGORI 1: HR BASIC (Pemanasan & Motivasi)
        # ==========================================
        {
            "role_family": role_family,
            "competency": "Career Motivation & Fit",
            "question_seed": f"Halo, selamat datang di sesi wawancara. Bisa ceritakan sedikit tentang dirimu dan kenapa kamu tertarik melamar posisi {target_role}?",
            "question_type": "hr_basic",
            "guardrail_signals": [
                "Menjelaskan latar belakang pendidikan atau pengalaman yang relevan",
                "Menyebutkan alasan atau motivasi spesifik memilih jalur karir ini",
                "Menunjukkan komunikasi yang percaya diri dan terstruktur"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Career Motivation & Fit",
            "question_seed": "Dari semua project atau materi yang pernah kamu pelajari, bagian mana sih yang bikin kamu merasa paling 'klik' dan yakin berkarir di jalur ini?",
            "question_type": "hr_basic",
            "guardrail_signals": [
                "Menyebutkan contoh pengalaman masa lalu atau proyek yang relevan",
                "Mengkorelasikan skill yang dimiliki dengan kebutuhan industri",
                "Menunjukkan tingkat antusiasme dan minat yang tinggi"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Career Motivation & Fit",
            "question_seed": f"Di industri IT perkembangannya kan cepat banget ya. Sebagai seorang {target_role}, bagaimana cara kamu biasanya belajar atau beradaptasi dengan teknologi baru?",
            "question_type": "hr_basic",
            "guardrail_signals": [
                "Menyebutkan metode belajar (misal: baca dokumentasi, ikut bootcamp, ngulik project)",
                "Menunjukkan kemauan belajar mandiri (self-taught ability)",
                "Memberikan contoh teknologi atau tools terbaru yang sedang dipelajari"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Career Motivation & Fit",
            "question_seed": f"Kalau kamu diterima di posisi {target_role} ini, apa target atau harapan terbesarmu dalam 1 sampai 2 tahun ke depan?",
            "question_type": "hr_basic",
            "guardrail_signals": [
                "Menjelaskan visi karir yang logis dan realistis",
                "Menunjukkan ambisi untuk berkontribusi lebih pada perusahaan",
                "Menyelaraskan tujuan pribadi dengan peran yang dilamar"
            ]
        },

        # ==========================================
        # KATEGORI 2: BEHAVIORAL (Soft Skills & STAR)
        # ==========================================
        {
            "role_family": role_family,
            "competency": "Teamwork & Collaboration",
            "question_seed": "Ceritakan pengalamanmu saat harus bekerja sama dalam tim untuk menyelesaikan sebuah proyek IT. Apa peran spesifikmu dan bagaimana kamu menangani perbedaan pendapat jika ada?",
            "question_type": "behavioral",
            "guardrail_signals": [
                "Menjelaskan situasi proyek dan pembagian peran dalam tim (Situation/Task)",
                "Menyebutkan tindakan komunikasi atau negosiasi yang dilakukan saat ada konflik (Action)",
                "Menjelaskan hasil akhir proyek dan penyelesaian masalah (Result)"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Adaptability & Problem Solving",
            "question_seed": "Pernahkah kamu menghadapi situasi di mana spesifikasi proyek tiba-tiba berubah padahal tenggat waktu sudah dekat? Ceritakan bagaimana kamu merespons dan menyelesaikannya.",
            "question_type": "behavioral",
            "guardrail_signals": [
                "Menjelaskan konteks perubahan mendadak dan dampaknya pada tenggat waktu (Situation/Task)",
                "Menyebutkan tindakan adaptasi, penentuan prioritas ulang, atau manajemen stres (Action)",
                "Menjelaskan hasil operasional atau pelajaran yang diambil dari kejadian tersebut (Result)"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Resilience & Handling Failure",
            "question_seed": "Di dunia IT, bikin error atau sistem nge-bug itu hal biasa. Coba ceritakan satu pengalamanmu saat melakukan kesalahan cukup fatal di sebuah project. Apa yang kamu lakukan untuk memperbaikinya?",
            "question_type": "behavioral",
            "guardrail_signals": [
                "Mengakui kesalahan secara terbuka tanpa menyalahkan orang lain (Situation/Task)",
                "Menyebutkan tindakan investigasi, perbaikan cepat (hotfix), atau koordinasi tim (Action)",
                "Menjelaskan lesson learned atau langkah pencegahan agar tidak terulang (Result)"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Communication & Empathy",
            "question_seed": "Ceritakan momen di mana kamu harus menjelaskan istilah teknis yang sangat rumit kepada orang awam (seperti dosen penguji, klien, atau teman beda jurusan). Bagaimana caramu memastikannya paham?",
            "question_type": "behavioral",
            "guardrail_signals": [
                "Menjelaskan konteks lawan bicara yang non-teknis (Situation/Task)",
                "Menyebutkan penggunaan analogi, bahasa sederhana, atau visualisasi (Action)",
                "Menjelaskan keberhasilan komunikasi dan pemahaman lawan bicara (Result)"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Initiative & Ownership",
            "question_seed": "Pernahkah kamu berinisiatif memperbaiki kode, sistem, atau proses yang sebenarnya bukan tugas utamamu? Ceritakan alasannya dan apa dampak dari inisiatifmu itu.",
            "question_type": "behavioral",
            "guardrail_signals": [
                "Menjelaskan penemuan masalah di luar tanggung jawab utama (Situation/Task)",
                "Menyebutkan tindakan sukarela mengambil alih atau mengusulkan solusi (Action)",
                "Menjelaskan dampak positif inisiatif tersebut bagi efisiensi tim atau produk (Result)"
            ]
        },

        # ==========================================
        # KATEGORI 3: CLOSING & SELF-AWARENESS
        # ==========================================
        {
            "role_family": role_family,
            "competency": "Self-Awareness & Growth Mindset",
            "question_seed": f"Sebagai seorang junior {target_role}, area teknis apa yang menurutmu masih menjadi kelemahan terbesarmu saat ini, dan apa rencanamu untuk memperbaikinya?",
            "question_type": "closing",
            "guardrail_signals": [
                "Menyebutkan kelemahan teknis spesifik secara jujur (tanpa terdengar fatal)",
                "Menjelaskan langkah konkrit atau rencana pembelajaran (seperti ikut kursus, baca dokumentasi)",
                "Menunjukkan kemauan belajar (growth mindset) yang kuat"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Self-Awareness & Culture Fit",
            "question_seed": "Lingkungan kerja atau dinamika tim seperti apa sih yang menurutmu paling bisa bikin kamu bekerja secara maksimal?",
            "question_type": "closing",
            "guardrail_signals": [
                "Menyebutkan preferensi gaya kerja (misal: kolaboratif, agile, komunikasi terbuka)",
                "Menunjukkan kesiapan untuk menerima masukan (feedback)",
                "Memberikan alasan logis yang selaras dengan nilai-nilai profesional"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Self-Awareness & Growth Mindset",
            "question_seed": "Kalau nanti kamu mulai bekerja di sini, menurutmu apa tantangan terbesar yang bakal kamu hadapi di 3 bulan pertamamu?",
            "question_type": "closing",
            "guardrail_signals": [
                "Menunjukkan pemahaman terhadap kompleksitas peran secara realistis",
                "Menyebutkan proses adaptasi teknis atau budaya kerja sebagai tantangan",
                "Memberikan sinyal bahwa kandidat sudah memikirkan strategi mitigasinya"
            ]
        },
        {
            "role_family": role_family,
            "competency": "Professionalism & Curiosity",
            "question_seed": "Wawancaranya sudah hampir selesai nih. Terakhir dari saya, apakah ada pertanyaan yang ingin kamu ajukan terkait posisi ini atau budaya kerja di perusahaan kami?",
            "question_type": "closing",
            "guardrail_signals": [
                "Mengajukan pertanyaan yang relevan dan kritis terkait operasional perusahaan",
                "Menunjukkan keingintahuan terhadap ekspektasi kinerja atau ekosistem teknologi",
                "Menutup sesi wawancara dengan ucapan terima kasih dan sikap profesional"
            ]
        }
    ]
```
