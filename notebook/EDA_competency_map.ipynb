```python
import pandas as pd
import json
```

```python
# 1. Buka dan baca file JSON-nya
with open('C:\\Users\\nurul\\OneDrive\\Desktop\\data-science\\data\\competency_map.json', 'r') as file:
    data_json = json.load(file)
```

```python
rows = []
```

```python
# 3. Looping untuk bongkar isi JSON
for role_name, role_info in data_json.items():
    domain = role_info.get('domain')
    role_family = role_info.get('role_family')
    
    # Masuk ke dalam list competencies
    for comp in role_info.get('competencies', []):
        rows.append({
            'target_role': role_name,
            'domain': domain,
            'role_family': role_family,
            'competency_id': comp.get('competency_id'),
            'competency_name': comp.get('competency'),
            'weight': comp.get('weight'),
            'total_skills': len(comp.get('related_skills', [])),
            'related_skills': ", ".join(comp.get('related_skills', []))
        })
```

```python
# 4. Ubah jadi Pandas DataFrame
df_comp = pd.DataFrame(rows)
```
