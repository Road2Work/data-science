```python
import json
import pandas as pd
import os
```

```python
# 1. Membaca file JSON
file_path = '../data/question_seed.json' 
with open(file_path, 'r') as f:
    data = json.load(f)
```

```python
# 2. Flatten data dan Membuat DataFrame
rows = []
for role_name, questions in data.items():
    for q in questions:
        rows.append({
            'target_role': role_name,
            'competency': q.get('competency'),
            'question_type': q.get('question_type')
        })

df = pd.DataFrame(rows)
df.head()
```

```python
# 3. Analisis Distribusi (Count per Role per Type)
summary = df.groupby(['target_role', 'question_type']).size().unstack(fill_value=0)

# 4. Tambahkan kolom Total & Validasi Ketentuan
summary['Total'] = summary.sum(axis=1)

print("--- Ringkasan Distribusi Pertanyaan per Role ---")
print(summary)
```
